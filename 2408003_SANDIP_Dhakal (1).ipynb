Outlier Detection:
Tasks:
1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
2. Define outliers using the 1.5 × IQR rule.
3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
in a different color.
4. Discuss the characteristics of these outliers and their potential impact on regional averages.

```python
import pandas as pd

# Calculate the IQR for Score and GDP per Capita Qno22

# For Score
Q1_score = SouthAsia['score'].quantile(0.25)
Q3_score = SouthAsia['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score
lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score

# For GDP per Capita
Q1_gdp = SouthAsia['Log GDP per capita'].quantile(0.25)
Q3_gdp = SouthAsia['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp
lower_bound_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_bound_gdp = Q3_gdp + 1.5 * IQR_gdp

# Identify outliers
outliers_score = SouthAsia[(SouthAsia['score'] < lower_bound_score) | (SouthAsia['score'] > upper_bound_score)]
outliers_gdp = SouthAsia[(SouthAsia['Log GDP per capita'] < lower_bound_gdp) | (SouthAsia['Log GDP per capita'] > upper_bound_gdp)]

# Combine outliers from both Score and GDP per Capita
outliers = pd.concat([outliers_score, outliers_gdp]).drop_duplicates()

print("Outliers based on Score and GDP per Capita:")
print(outliers)
```

```python
import matplotlib.pyplot as plt

# Scatter plot with GDP per Capita on x-axis and Score on y-axis Qno 23
plt.figure(figsize=(10, 6))

# Plot all countries in one color (e.g., blue)
plt.scatter(SouthAsia['Log GDP per capita'], SouthAsia['score'], color='blue', label='Normal', alpha=0.6)

# Highlight outliers in a different color (e.g., red)
plt.scatter(outliers['Log GDP per capita'], outliers['score'], color='red', label='Outliers', alpha=1)

# Labels and title
plt.xlabel('Log GDP per Capita')
plt.ylabel('Score')
plt.title('Scatter Plot of Score vs. GDP per Capita with Outliers Highlighted')
plt.legend()

# Show the plot
plt.show()
```
